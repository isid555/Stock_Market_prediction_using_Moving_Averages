### Stock Market Prediction using Moving Averages

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as  plt 
import pandas_datareader as data 
import yfinance as yf 
```

```python
from datetime import datetime

current_date = datetime.now().strftime('%Y-%m-%d')
print(current_date)
```

```python
start ='2013-05-05'
end = current_date
stock = 'BPCL.NS'
df = yf.download(stock, start=start, end=end)
print(df.tail())
```

```python
df.shape
```

```python
plt.plot(df.Close)
```

##### Creating 100 & 200 days moving averages

```python
ma100  = df.Close.rolling(100).mean()
ma200 = df.Close.rolling(200).mean()
plt.figure(figsize=(15,9))
plt.plot(df.Close)
plt.plot(ma100, 'r')
plt.plot(ma200 , 'g')
plt.title("Stock Prediction using Moving Averages")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.show()
```

When the stock price crosses above the 100-day moving average from below, it may indicate a potential bullish trend.
Conversely, when the stock price crosses below the 100-day moving average from above, it may suggest a potential bearish trend.
The 200-day moving average is often used for longer-term trends, and crossovers with the stock price can signal more significant shifts.

# Disclaimer
Remember that these signals are not foolproof, and false signals can occur. Additionally, incorporating other technical indicators, fundamental analysis, and market sentiment can enhance the robustness of your stock predictions. Always be cautious and consider using multiple tools for analysis when making financial decisions.
